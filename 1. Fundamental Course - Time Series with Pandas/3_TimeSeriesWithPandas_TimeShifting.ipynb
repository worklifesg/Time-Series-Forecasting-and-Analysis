<h2 align='center'> Time Series with Pandas</h2>

<h3> In this program we will be working on Time Shifting </h3>

Time shifting --> Sometimes we need shift the data up/down along time series index and pandas has in-built function for this funcationality

```python
# Import libraries
from datetime import datetime
import numpy as np
import pandas as pd
```

<h3> Read Dataset (Import Dataset)</h3>

```python
#read dataset 'starbucks.csv'
df = pd.read_csv('starbucks.csv',index_col='Date',parse_dates=True)
df.head()
```

```python
df.tail()
```

<h4> Now we need to shift entire day index a given number of rows without regrad for periods such as months or years.

```python
df.shift(1) #shift all data points for whole row, here we shift the data by 1 row and the shifted row is NaN missing value)
# we lose last datat point and get an empty one in the beginning
```

To permanently affect your dataframe, we can assign df = df.shift(1) and data points are remvoed permanently

```python
df.shift(-1) #shifting backwards
# Now the empty point is created at the bottom
```

```python
df.head()
```

```python
# Shift everything forward for one month (end of the month)

df.shift(periods=1,freq='M').head()
```

```python
# Using fill_value to put something in new created matrix
#Using fill value as mean of Close Value

df.shift(periods=1,fill_value=df['Close'].mean()).head()
```
